```python
#IMPORT THE LIBRARIES
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.preprocessing import MinMaxScaler,StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import TimeSeriesSplit
import warnings
warnings.filterwarnings('ignore')
```

```python
#step 2 :FEATURE ENGINEERING
print('\n'+'--'*40)
print("FEATURE ENGINERING")
print('--'*40)
```

```python
data=df.copy()

#--moving averages
for w in [5,7,10,21,50,100,200]:
    data[f'MA_{w}']=data['Close'].rolling(w).mean()
#--exponential movingg average
for s in [9,12,20,26]:
    data[f'EMA_{s}']=data['Close'].ewm(span=s,adjust=False).mean()
#--MACD
data['MACD']=data['EMA_12']-data['EMA_26']
data['MACD_Signal']=data['MACD'].ewm(span=9,adjust=False).mean()
data['MACD_Hist']=data['MACD']-data['MACD_Signal']
data['MACD_Crossover']=(data['MACD']>data['MACD_Signal']).astype(int)

#---RSI
delta=data['Close'].diff()
gain=delta.clip(lower=0).rolling(14).mean()
loss=(-delta.clip(upper=0)).rolling(14).mean()
data['RSI']=100-(100/(1+ gain/(loss+1e-10)))
data['RSI_Overbought']=(data['RSI']>70).astype(int)
data['RSI_Oversold']=(data['RSI']<30).astype(int)

#---Bollinger Bands
data['BB_Mid']=data['Close'].rolling(20).mean()
data['BB_Std']=data['Close'].rolling(20).std()
data['BB_Upper']=data['BB_Mid']+2*data['BB_Std']
data['BB_Lower']=data['BB_Mid']-2*data['BB_Std']
data['BB_Width']=(data['BB_Upper']-data['BB_Lower'])/(data['BB_Mid']+1e-10)
data['BB_Pct']=(data['Close']-data['BB_Lower'])/(data['BB_Upper']-data['BB_Lower']+1e-10)

#---Target the next day close
data['Target_1d']=data['Close'].shift(-1)
data['Target_5d']=data['Close'].shift(-5)
data['Target_21d']=data['Close'].shift(-21)

data.dropna(inplace=True)
data.reset_index(drop=True,inplace=True)
```

```python
#step 3:prepare features
print('\n'+'--'*40)
print('PREPARE FEATURES')
print('--'*40)
```

```python
feature_cols=[c for c in data.columns if c not in 
             ['Date','Open','High','Low','Volume','Target_1d','Target_5d','Target_21d',
             'BB_Std','BB_Mid']]
X=data[feature_cols]
y=data['Target_1d']  #forecast ext days close

#---TIME SERIES SPLIT--- NEVER SHUFFLE FINANCIAL DAT!!!
split=int(len(data)*0.8)
X_train,X_test=X.iloc[:split],X.iloc[split:]
y_train,y_test=y.iloc[:split],y.iloc[split:]
close_test =data['Close'].iloc[split:].values

scaler=StandardScaler()
X_train_s=scaler.fit_transform(X_train)
X_test_s=scaler.transform(X_test)

print(f"\n Training period:{data['Date'].iloc[0]}-->{data['Date'].iloc[split-1]}({split:,}days)")
print(f"\nTest period: {data['Date'].iloc[split]}-->{data['Date'].iloc[-1]}({len(X_test):,}days)")
print(f"Features used: {len(feature_cols)}")
```

```python
#---------------------------------------------------------------
#step 4:train an evaluate models
#---------------------------------------------------------------

print('\n'+'--'*40)
print('TRAINING &EVALUATING 5 MODELS')
print('--'*40)
```

```python
models={
    'Linear Regression': LinearRegression(),
    'Ridge Regression':  Ridge(alpha=10.0),
    'Lasso Regression':  Lasso(alpha=0.1,max_iter=5000),
    'Random Forest'   :  RandomForestRegressor(n_estimators=300,max_depth=4,
                                               min_samples_split=10,
                                               random_state=42,n_jobs=-1),
    'Gradient Boosting': GradientBoostingRegressor(n_estimators=300,max_depth=4,
                                                  learning_rate=0.03,subsample=0.8,
                                                  random_state=42),
    
}
results={}
tscv=TimeSeriesSplit(n_splits=5)
print(f"\n{'Model':<22}{'RSME':>8}{'MAE':>7}{'MAPE':>7}{'DirAcc':>8}{'CV AUC':>8}")
print(f"{'--'*22}{'--'*8}{'--'*7}{'--'*8}{'--'*8}")

for  name, model in models.items():
    #cross validation
    cv_scores=[]
    for tr_idx,val_idx in tscv.split(X_train_s):
        model.fit(X_train_s[tr_idx],y_train.iloc[tr_idx])
        val_pred=model.predict(X_train_s[val_idx])
        val_rmse=np.sqrt(mean_squared_error(y_train.iloc[val_idx],val_pred))
        cv_scores.append(val_rmse)

        #final fit &and TEst
        model.fit(X_train_s,y_train)
        preds=model.predict(X_test_s)
        rmse=np.sqrt(mean_squared_error(y_test,preds))
        mae=mean_absolute_error(y_test,preds)
        mape=np.mean(np.abs((y_test.values-preds)/(y_test.values+1e-10)))*100

        #directional accuracy
        actual_dir=np.sign(y_test.values--close_test)
        pred_dir=np.sign(preds-close_test)
        dir_acc=(actual_dir==pred_dir).mean()*100
        cv_mean=np.mean(cv_scores)
        results[name]={'model':model,'preds':preds,
                      'rmse':rmse,'mae':mae,'mape':mape,
                      'dir_acc':dir_acc,'cv_rmse':cv_mean}
        print(f"{name:<22}${rmse:>6.2f}${mae:>6.2f}{mape:>5.2f}%{dir_acc:>6.1f}%${cv_mean:>6.2f}")

        #best modekls by rmse
        best_name=min(results,key=lambda x:results[x]['rmse'])
        best=results[best_name]
        best_model=best['model']

        print(f"\n best model:  {best_name}")
        print(f"RMSE:${best['rmse']:.2F}AVG ERROR PER SHARE")
        print(f"MAE:${best['mae']:.2f}")
        print(f"mape:{best['mape']:.2f}%")
        print(f"Directional acc:{best['dir_acc']:.1f}")
```

```python
#-------------------------------------------------------
#step 5:backtest-simple trading strategy
#-------------------------------------------------------

print('\n'+'--'*40)
print("BACKTESTING- SIMPLE ML TRADING STRATEGY")
print('--'*40)
```

```python
preds=best['preds']
cash=10000 #start with 10000
shares=0
portfolio=[]
buy_prices=[]
trades=[]

for i in range(len(preds)):
    price_now=close_test[i]
    price_pred=preds[i]
    pred_return=(price_pred-price_now)/price_now*100

    if pred_return>2.0 and cash>=price_now and shares==0:
        #BUY SIGNAL
        shares=int(cash/price_now)
        cash-=shares*price_now
        buy_prices.append(price_now)
        trades.append(('BUY',i,price_now,shares))

    elif(pred_return<-1.5 or i==len(preds)-1) and shares>0:
        #sell SIGNAL
        sell_val=shares*price_now
        profit=sel_val-(shares*buy_prices[-1])
        cash+=sel_val
        trades.append(('SELL',i,price_now,shares))
        shares=0

    portfolio_val=cash+shares*price_now
    portfolio.append(portfolio_val)

final_val=portfolio[-1] if portfolio else 1000
total_return=(final_val-10000)/10000*100

#buy &hold comparison
bh_shares=int(10000/close_test[0])
bh_final=bh_shares*close_test[-1]+(10000-bh_shares*close_test[0])
bh_return=(bh_final-10000)/10000*100

buy_trades=[t for t in trades if t[0]=='BUY']
sell_trades=[t for t in trades if t[0]=='SELL']

print(f'\nStarting capital:$10000')
print(f"Final portfolio:${final_val:,.2f}")
print(f"ML strategy return :{total_return:+.1f}%")
print(f"buy &and hold return :{bh_return:+.1f}%")
print(f"total shares:    {len(buy_trades)}buys,{len(sell_trades)} sells")

if buy_trades:
    print(f"\nSample TRades:")
    print(f"{'Action':<6}{'Day':>5}{'Price':>9}{'Shares':>8}")
    
```
